```python

def run_casscf_calculation():
    # Define scanning properties for y and z directions
    input = {
    "molecule": "helium",
    "method": "CASSCF(2,2)",
    "basis": 'cc-pVDZ',
    "max_workers": 6,
    "grid": {'x':(0,0.0), 'y':(0,0.0), 'z':(1.0,0.01)},
    "properties": ['pair_density_nucleus',"pair_density_nucleus_c1", "pair_density_nucleus_c2", "pair_density_nucleus_hf"
                   , 'intracule', 'intracule_c1', 'intracule_c2', 'intracule_hf'],
}

    # Run calculation for CASSCF(2,2) with scanning properties
    results = run_calculation(input)

    return results
```

```python
results = run_casscf_calculation()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def plot_pair_density_vs_z(data_path=None):
    """
    Plot pair_density_nucleus_c1 and pair_density_nucleus_c2 vs z-coordinate.

    Parameters:
    -----------
    data_path : str or pandas.DataFrame, optional
        Path to data file (CSV, Excel, etc.) or pandas DataFrame containing 'z',
        'pair_density_nucleus_c1', and 'pair_density_nucleus_c2' columns.
        If None, sample data will be generated.

    Returns:
    --------
    fig : matplotlib.figure.Figure
        The matplotlib figure object
    """
    # Load data from file if provided
    if data_path is not None:
        if isinstance(data_path, pd.DataFrame):
            df = data_path
            # Filter for z >= 0
            df = df[df['z'] >= 0]
        elif data_path.endswith('.csv'):
            df = pd.read_csv(data_path)
        elif data_path.endswith(('.xls', '.xlsx')):
            df = pd.read_excel(data_path)
        else:
            raise ValueError("Unsupported file format. Please provide a CSV or Excel file.")
    else:
        # If no data is provided, create sample data
        z = np.linspace(-5, 5, 100)
        # Sample data with different profiles for C1 and C2
        c1 = 0.5 * np.exp(-0.3 * z**2)
        c2 = 0.3 * np.exp(-0.5 * (z-1)**2)

        df = pd.DataFrame({
            'z': z,
            'pair_density_nucleus_c1': c1,
            'pair_density_nucleus_c2': c2
        })

    # Create the plot
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plot pair_density_nucleus_c1 vs z

#    ax.plot(df['z'], df['pair_density_nucleus'], '-',
#            color='black', linewidth=3, markersize=4,
#            label='pair_density_nucleus')

    ax.plot(df['z'], df['pair_density_nucleus_c1'], '-',
            color='#077FAE', linewidth=2, markersize=4,
            label='pair_density_nucleus_c1')

    # Plot pair_density_nucleus_c2 vs z
    ax.plot(df['z'], df['pair_density_nucleus_c2'], '-',
            color='#C29044', linewidth=2, markersize=4,
            label='pair_density_nucleus_c2')

  #  ax.plot(df['z'], df['intracule'], '--',
   #         color='black', linewidth=3, markersize=4,
   #         label='intracule')

    ax.plot(df['z'], df['intracule_c1'], '--',
            color='#077FAE', linewidth=2, markersize=4,
            label='intracule_c1')

    ax.plot(df['z'], df['intracule_c2'], '--',
            color='#C29044', linewidth=2, markersize=4,
            label='intracule_c2')



    # Customize the plot
    ax.set_xlabel('z coordinate (Å)', fontsize=12)
    ax.set_ylabel('Property', fontsize=12)
    ax.set_title('Pair Density and intraculue Components vs Z Coordinate', fontsize=14, fontweight='bold')
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()



    plt.tight_layout()

    return fig

# Example usage
if __name__ == "__main__":
    # Example with sample data
    fig = plot_pair_density_vs_z(results['cube_df'])
    plt.show()

    # Example with provided dataframe
    # plot_pair_density_vs_z(results['cube_df'])
```
