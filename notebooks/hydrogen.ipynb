```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def plot_z_vs_xr(data_path, molec):
    """
    Plot z-coordinate vs X(r) from either a provided dataframe/path or sample data.

    Parameters:
    -----------
    data_path : str or pandas.DataFrame, optional
        Path to data file (CSV, Excel, etc.) or pandas DataFrame containing 'x' and 'X(r)' columns.
        If None, sample data will be generated.

    Returns:
    --------
    fig : matplotlib.figure.Figure
        The matplotlib figure object
    """
    # Load data from file if provided
    if data_path is not None:
        if isinstance(data_path, pd.DataFrame):
            df = data_path
            df['x'] = df['x']*0.529177249
        elif data_path.endswith('.csv'):
            df = pd.read_csv(data_path)
        elif data_path.endswith(('.xls', '.xlsx')):
            df = pd.read_excel(data_path)
        else:
            raise ValueError("Unsupported file format. Please provide a CSV or Excel file.")


    # Create the plot
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plot X(r) vs z
    ax.plot(df['x'], df['X(r)'], 'o-', color='#077FAE', linewidth=2, markersize=4, label='X(r)')

    # Add reference line at X(r) = 1
    ax.axhline(y=1, color='gray', linestyle='--', alpha=0.7, label='Reference (X(r) = 1)')

    # Customize the plot
    ax.set_xlabel('z coordinate (Å)', fontsize=12)
    ax.set_ylabel('X(r)', fontsize=12)
    ax.set_title(f'Hydrogen{molec} CASSCF(2,2) TZVP', fontsize=14, fontweight='bold')
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()

    # Add text explaining the significance of X(r)


    plt.tight_layout()
    plt.show()
    return fig

#df = results_hyd['cube_df']
#plot_z_vs_xr(df, '')
```

```python
    # Define scanning properties for y and z directions
input = {
    "molecule": "hydrogen_2eq",
    "method": "CASSCF(2,2)",
    "basis": 'TZVP',
    "config": 'SP',
    "max_workers": 6,
    "grid": {'x':(0,0), 'y':(0,0), 'z':(4,0.05)},
    "properties": ['otr']
}

    # Run calculation for CASSCF(2,2) with scanning properties
results_hyd2 = run_calculation(input)
df = results_hyd2['cube_df']
plot_z_vs_xr(df, '(2eq)')
```
